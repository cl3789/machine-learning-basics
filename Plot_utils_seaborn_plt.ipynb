```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interactive, interact

import seaborn as sns
%matplotlib inline

from sklearn import datasets

import altair as alt
```

```python
iris_df = datasets.load_iris(as_frame = True)

X = iris_df.data
y = pd.DataFrame({"label":iris_df.target})
```

```python
## Concatenate X and y horizontally
iris_df = pd.concat([X,y], axis=1)

## Rename columns
iris_df = iris_df.rename(columns={"sepal length (cm)": "sepal_length",
                       "sepal width (cm)": "sepal_width",
                       "petal length (cm)": "petal_length",
                       "petal width (cm)": "petal_width"})


## Load mnist data
mnist = datasets.load_digits()
X_mnist = mnist.data
y_mnist = mnist.target
```

```python
def plot_scatter(data, 
                 x_var:str, # x-axis variable column name
                 y_var:str, # y-axis variable column name
                 class_label:str, # class label column name
                 class_dict:dict, # dictionary maps the class name (numeric) to string
                 palette:str, # color palette 
                 hue_order:list, # order for classes
                 point_size:int, # point size
                 title:str, # main title
                 x_label:str, # x-axis label
                 y_label:str, # y-axis label
                 x_lim:list,  # x-axis range
                 y_lim:list,  # y-axis range
                 offset):
    """2D Scatter plot of selected features"""
    plt.figure(figsize = (6, 6))

    data_df = data

    data_df["class_str"] = data_df[class_label].map(class_dict)

    sns.set_context("notebook")
    sns.set_style("ticks")
    sns.scatterplot(data=data_df, 
                    x=x_var, 
                    y=y_var, 
                    hue="class_str",
                    palette=palette,
                    hue_order = hue_order,
                    legend=True,
                    s = point_size,
                    alpha = 0.5)
    sns.despine(offset=offset)

    plt.gca().set(
        aspect="equal",
        title=title,
        xlabel=x_label,
        ylabel=y_label,
        xlim = x_lim,
        ylim = y_lim
    )

    plt.legend(loc='center left', title="label", bbox_to_anchor=(1, 0.5))
    plt.show()

```

```python
x_var = "sepal_length"
y_var = "sepal_width"
x_label = "sepal length (cm)"
y_label = "sepal width (cm)"
class_label = "label"
class_dict = {0: "setosa", 1: "versicolor", 2:"virginica"}
hue_order = ["setosa", "versicolor",  "virginica"]
x_lim = [0,10]
y_lim = [0,10]

plot_scatter(
             data = iris_df,
             x_var= x_var, 
             y_var= y_var,
             class_label= class_label,
             class_dict= class_dict,
             palette= "deep",
             hue_order= hue_order,
             point_size=100,
             title=f"Scatter plot of {x_var} vs. {y_var}",
             x_label=x_label,
             y_label=y_label,
             x_lim = x_lim,
             y_lim = y_lim,
             offset = 10)
```

```python
def plot_subplot(img_list, n_row, n_col, width, height, show_label=True):
    """Create (x_row, n_col) subplot of a list of images"""
    px = 1/plt.rcParams['figure.dpi']  # pixel in inches
    fig, axes = plt.subplots(nrows=n_row, ncols=n_col, figsize=(width, height))
    plt.subplots_adjust(hspace = .001)

    count = -1 
    for i in range(n_row):
        for j in range(n_col):
            count = count+1
            axes[i,j].imshow(img_list[count]) ## show image
            axes[i,j].grid(False)
            axes[i,j].set_xticks([])
            axes[i,j].set_yticks([])
            
    rows = ['Row label {}'.format(row) for row in range(n_row)]    ## labels on the Rows
    cols = ['Column label {}'.format(col) for col in range(n_col)] ## labels on the Columns

    if show_label:
        rows = ['Row label {}'.format(row) for row in range(n_row)]    ## labels on the Rows
        cols = ['Column label {}'.format(col) for col in range(n_col)] ## labels on the Columns
        for ax, col in zip(axes[0], cols):
            ax.set_title(col, size='small')

        for ax, row in zip(axes[:,0], rows):
            ax.set_ylabel(row, rotation=90, size='small')

    fig.tight_layout()
    plt.show()  
```

```python
## Select first 5 samples from each digit.
selected_digits = []
k = 9 # select k samples per digit label
for n in range(10):
    y_mnist_n = (y_mnist == n)       # create boolean mask based on digit label. select digit=n
    X_mnist_n = X_mnist[y_mnist_n,:] # select corresponding Row in X_mnist with desired digit label
    X_mnist_n = list(X_mnist_n[0:k].reshape(k, 8, 8))  # select first k sample
    selected_digits.extend(X_mnist_n) # appending all sample in one list
    
plot_subplot(selected_digits, 10, 9, width=8, height=8, show_label = False)
```
